```python
import cv2
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("kitten.jpg")
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img, 200, 300)
plt.imshow(edges)
```

```python
plt.imshow(edges, plt.cm.gray)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGRA2RGBA)
plt.imshow(gray)
```

```python
grayed = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grayed)
```

```python
inverted_gray = 255 - grayed
plt.imshow(inverted_gray)
```

```python
blur = cv2.GaussianBlur(inverted_gray, (21, 21), 0)
plt.imshow(blur)
```

```python
inverted_blur = 255 - blur
plt.imshow(inverted_blur)
```

```python
pencil_sketch = cv2.divide(grayed, inverted_blur, scale=256.0)
plt.imshow(pencil_sketch)
```

```python
inverted = 255 - img
blur_img = cv2.GaussianBlur(img, (21,21), 0)
inverted_blur_img = 255 - blur_img
sketch_img = cv2.divide(img, inverted_blur_img, scale=256.0)
plt.imshow(sketch_img)
```
