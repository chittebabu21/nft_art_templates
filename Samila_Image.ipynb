```python
import random
import math
import matplotlib.pyplot as plt
from samila import GenerativeImage, Projection
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (16, 12)
```

```python
def f1(x, y):
    result = random.uniform(-1, 1) * x**2 - math.sin(y**2) + abs(x-y)
    return result
```

```python
def f2(x, y):
    result = random.uniform(-1, 1) * y**3 - math.cos(x**3) + 2*x
    return result
```

```python
g = GenerativeImage(f1, f2)
```

```python
g.generate()
```

```python
g.seed
```

```python
g.plot(projection=Projection.LAMBERT, color="aqua", bgcolor="black")
```
